```python
#feature enginieering
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info() 
```

```python
# new df created is still in object(string) need to convert it into a int
df['Date'] = df['Date'].astype('int64')
df['Month'] = df['Month'].astype('int64')
df['Year'] = df['Year'].astype('int64')
```

```python
df.info()
```

```python
# Dropping Date_of_Journey  column

df.drop('Date_of_Journey', axis = 1, inplace = True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0]) # to get rid of the date which was coming in as well
```

```python
df['Arrival_Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df.info()
```

```python
# convering to int 
df['Arrival_Hour'] = df['Arrival_Hour'].astype('int64')
df['Arrival_Minute'] = df['Arrival_Minute'].astype('int64')
```

```python
# dropping Arrival_time
df.drop('Arrival_Time', axis = 1, inplace = True)
```

#### Univariate Analysis

```python
# distribution of flight prices
plt.figure(figsize=(8,5))
sns.histplot(df['Price'], kde=True)
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Frequenvy of airline apperaing the most
df['Airline'].value_counts().plot(kind='bar', figsize=(10,4), color='skyblue')
plt.title('Count of Flights per Airline')
plt.xlabel('Airline')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# common source and destination
df['Source'].value_counts().plot(kind='bar', title='Source City Count')
plt.show()

df['Destination'].value_counts().plot(kind='bar', title='Destination City Count')
plt.show()
```

#### Multivariate Analysis 


```python
#Flight price vary by airline
plt.figure(figsize=(12,5))
sns.boxplot(x='Airline', y='Price', data=df)
plt.xticks(rotation=45)
plt.title('Flight Price by Airline')
plt.show()
```

```python
# stops influence price
sns.boxplot(x='Total_Stops', y='Price', data=df)
plt.title('Price vs Number of Stops')
plt.show()
```
